```python
#Import Libraries
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
dataset=pd.read_csv(r"../../Data Science/Cursos Online/Udemy/Machine-Learning-A-Z-New/Machine Learning A-Z New" \
                    "/Part 4 - Clustering/Section 24 - K-Means Clustering/Mall_Customers.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,3:5].values
y=  dataset.iloc[:,-1].values
plt.scatter(X[:,0], X[:,1])
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])
```

```python
K_model = AgglomerativeClustering(n_clusters=5, affinity='euclidean')
k_pred =  K_model.fit_predict(X)
```

```python
figure=plt.figure(figsize=(15,10))


plt.scatter(X[k_pred==0][:,0], X[k_pred==0][:,1], s=100, c='green', label = 'cls_1')
plt.scatter(X[k_pred==1][:,0], X[k_pred==1][:,1], s=100, c='blue', label = 'cls_2')
plt.scatter(X[k_pred==2][:,0], X[k_pred==2][:,1], s=100, c='red', label = 'cls_3')
plt.scatter(X[k_pred==3][:,0], X[k_pred==3][:,1], s=100, c='yellow', label = 'cls_4')
plt.scatter(X[k_pred==4][:,0], X[k_pred==4][:,1], s=100, c='magenta', label = 'cls_5')
# plt.scatter(X[k_pred==5][:,0], X[k_pred==5][:,1], s=100, c='purple', label = 'cls_6')
# plt.scatter(K_model.cluster_centers_[:,0], K_model.cluster_centers_[:,1], s=200, marker='^', c='red', label='centroids')
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])
plt.legend()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
figure=plt.figure(figsize=(15,10))

dendongram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendogram')
plt.xlabel('Costumers')
plt.ylabel('Euclidean Distances')
```
